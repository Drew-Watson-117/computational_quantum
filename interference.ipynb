```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x0 = -2
D = 5
L = 5
x = np.arange(x0, D, 0.01)
y = np.arange(-L, L, 0.01)
t = np.arange(0, 5, 0.1)
T, X, Y = np.meshgrid(t, x, y, indexing='ij')
X.shape
```

```python
def heaviside(x):
    return np.where(x >= 0, 1, 0)

def square(x, center, width):
    return heaviside(x - center + width/2) - heaviside(x - center - width/2)
```

```python
d = 0.5 # Distance between slits
a = 0.25 # Width of slits
tau = 0.5 # Thickness of the plate
V_0 = 10.0 # Potential height of the plate

V_x = square(X, 0, tau)
V_y = 1 - square(Y, -d/2, a) - square(Y, d/2, a)
V = V_0 * V_x * V_y
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X[0], Y[0], V[0])
plt.show()
```
